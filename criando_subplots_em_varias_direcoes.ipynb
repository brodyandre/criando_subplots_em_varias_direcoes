#📈 Análise de Vendas Mensais por Loja - 2022
###📋 Objetivo
Este relatório tem como objetivo analisar o desempenho mensal de vendas de quatro lojas (A, B, C e D) ao longo do ano de 2022, identificar padrões sazonais, avaliar performance individual e propor insights estratégicos.

###🔧 Preparação dos Dados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Definindo o nome das lojas
lojas = ['A', 'B', 'C', 'D']

# Dicionário contendo o número de vendas por mês para cada loja
vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

###📊 Visualização Gráfica

```python
# Criando o DataFrame com os dados
df_vendas = pd.DataFrame(vendas_2022, index=lojas)


fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

df_vendas.loc['A'].plot(ax=axes[0, 0], title='Loja A')
df_vendas.loc['B'].plot(ax=axes[0, 1], title='Loja B')
df_vendas.loc['C'].plot(ax=axes[1, 0], title='Loja C')
df_vendas.loc['D'].plot(ax=axes[1, 1], title='Loja D')

for ax in axes.flat:
    ax.set_xlabel('Mês')
    ax.set_ylabel('Número de Vendas')


ymin = 0
ymax =500
for ax in axes.ravel():
    ax.set_ylim(ymin, ymax)

plt.suptitle('Variação no Número de Vendas Mensais por Loja', fontsize=16)
plt.show()
```

##📌 Análises Relevantes
🔍 1. Sazonalidade Comum

* Todas as lojas apresentam aumento expressivo em novembro e dezembro.

* Indica influência de datas comerciais como Black Friday e Natal.

###✅ Loja C

* Melhor performance geral.

* Crescimento sólido e consistente.

* Destaque absoluto no fim do ano (🔺 Dezembro: 400 vendas).

* Potencial referência de boas práticas para outras lojas.

###⚠️ Loja A

* Bom desempenho até novembro.

* Apresenta queda significativa em dezembro (de 400 para 300 vendas), contrariando o padrão.

* Deve-se investigar problemas operacionais ou logísticos.

###⚠️ Loja B

* Crescimento moderado ao longo do ano.

* Dezembro foi o ponto alto (350 vendas), mas ainda abaixo da Loja C.

###🚨 Loja D

* Desempenho mais baixo entre todas.

* Mesmo com crescimento, seu pico foi de apenas 250 vendas.

* Precisa de revisão de estratégia, marketing, localização ou equipe

| Insight                            | Ação Recomendada                           |
| ---------------------------------- | ------------------------------------------ |
| 🔼 Alta sazonalidade no fim do ano | Planejar promoções e estoques para nov/dez |
| 🏅 Loja C com melhor performance   | Estudar processos, campanhas e equipe      |
| 📉 Queda na Loja A em dezembro     | Diagnosticar possíveis gargalos            |
| 📉 Baixa performance da Loja D     | Reavaliar modelo operacional               |
