#ðŸ“ˆ AnÃ¡lise de Vendas Mensais por Loja - 2022
###ðŸ“‹ Objetivo
Este relatÃ³rio tem como objetivo analisar o desempenho mensal de vendas de quatro lojas (A, B, C e D) ao longo do ano de 2022, identificar padrÃµes sazonais, avaliar performance individual e propor insights estratÃ©gicos.

###ðŸ”§ PreparaÃ§Ã£o dos Dados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Definindo o nome das lojas
lojas = ['A', 'B', 'C', 'D']

# DicionÃ¡rio contendo o nÃºmero de vendas por mÃªs para cada loja
vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

###ðŸ“Š VisualizaÃ§Ã£o GrÃ¡fica

```python
# Criando o DataFrame com os dados
df_vendas = pd.DataFrame(vendas_2022, index=lojas)


fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

df_vendas.loc['A'].plot(ax=axes[0, 0], title='Loja A')
df_vendas.loc['B'].plot(ax=axes[0, 1], title='Loja B')
df_vendas.loc['C'].plot(ax=axes[1, 0], title='Loja C')
df_vendas.loc['D'].plot(ax=axes[1, 1], title='Loja D')

for ax in axes.flat:
    ax.set_xlabel('MÃªs')
    ax.set_ylabel('NÃºmero de Vendas')


ymin = 0
ymax =500
for ax in axes.ravel():
    ax.set_ylim(ymin, ymax)

plt.suptitle('VariaÃ§Ã£o no NÃºmero de Vendas Mensais por Loja', fontsize=16)
plt.show()
```
